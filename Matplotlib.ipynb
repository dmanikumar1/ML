```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y,"g*")
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel("value")
plt.ylabel("cosine value")
plt.title("cos wave")
plt.show()
```

```python
plt.plot(x,z,"g-")
plt.plot(x,y,"r--")
plt.show()
```

```python
lang=["eng","tel","hin","french","Russian"]
people=[150,50,100,45,120]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(lang,people)
plt.xlabel("languages")
plt.ylabel("number of people")
plt.title("languages spoken")
plt.show()
```
